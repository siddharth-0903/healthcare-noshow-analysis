```python
import pandas as pd


```

```python
df = pd.read_csv("D:\Projects\healthcare-noshow-analysis\data\KaggleV2-May-2016.csv")
```

```python
df.info()
```

```python
df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay'])
```

```python
print("Shape:", df.shape)

print("\nData Types:\n", df.dtypes)

print("\nNull Values:\n", df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
df.rename(columns={
    'Hipertension': 'Hypertension',
    'Handcap': 'Handicap',
    'No-show': 'No_show',
    'ScheduledDay': 'Scheduled_Day',
    'AppointmentDay': 'Appointment_Day'
}, inplace=True)

# Make all column names lowercase
df.columns = df.columns.str.lower()

print(df.columns)
```

```python
df['days_between'] = (df['appointment_day'] - df['scheduled_day']).dt.days
```

```python
df.head()
```

```python
# Check how many negative values
print("Negative days:", (df['days_between'] < 0).sum())

# Remove them
df = df[df['days_between'] >= 0]
```

```python
print(df.shape)
print(df.head())
```

```python
df
```

```python
df['patientid'] = df['patientid'].astype(str)
df['appointmentid'] = df['appointmentid'].astype(str)

df.to_csv("cleaned_appointments.csv", index=False)
```

```python
df
```

```python

df.shape          # how many rows and columns
df.head()         # first 5 rows
df.dtypes         # column data types

```

```python
df.columns.tolist() 
```

```python
df['no_show'] = df['no_show'].map({'Yes': 1, 'No': 0})

# Verify
df['no_show'].value_counts()
df['no_show'].value_counts(normalize=True).round(3)  
```

```python
print(df['age'].min(), df['age'].max())
print(f"Negative ages: {(df['age'] < 0).sum()}")
print(f"Unrealistic ages: {(df['age'] > 115).sum()}")

# Remove rows with age below 0
df = df[df['age'] >= 0]
```

```python
print(df['gender'].value_counts())
print(df['handicap'].value_counts())   # should be 0-4, check for anything weird
print(df['scholarship'].value_counts()) # should only be 0 and 1
print(df['hypertension'].value_counts())
print(df['diabetes'].value_counts())
print(df['alcoholism'].value_counts())
print(df['sms_received'].value_counts())
```

```python
# convert to binary
df['handicap'] = df['handicap'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df['handicap'].value_counts()   # should be 0-4, check for anything weird
```

```python
df['appointment_weekday']     = df['appointment_day'].dt.day_name()
df['appointment_month']       = df['appointment_day'].dt.month_name()
df['appointment_weekday_num'] = df['appointment_day'].dt.dayofweek  # 0=Monday
df['scheduled_hour']          = df['scheduled_day'].dt.hour
```

```python
bins   = [-1, 17, 35, 60, 115]
labels = ['0-17', '18-35', '36-60', '60+']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

df['age_group'].value_counts()
```

```python
print("Shape after cleaning:", df.shape)
print("Nulls remaining:\n", df.isnull().sum())
print("No-show rate:", df['no_show'].mean().round(3))
print("Date range:", df['appointment_day'].min(), "to", df['appointment_day'].max())

df.dtypes
```

```python
df.head()
```

```python
df['patientid'] = df['patientid'].astype(str).str.replace('.0', '', regex=False)
df['appointmentid'] = df['appointmentid'].astype(str).str.replace('.0', '', regex=False)
```

```python
df.to_csv('cleaned_appointments.csv', index=False)
print(f"Saved {df.shape[0]} rows, {df.shape[1]} columns")
```


## Data Cleaning Summary
- Started with 110,527 rows, 14 columns
- Removed 385688 rows with negative days_between (scheduling errors) 
- Converted no_show to binary: 1 = no-show, 0 = showed up
- Fixed column names (corrected misspellings: Hipertension, Handcap)
- Engineered: days_between, age_group, appointment_weekday, appointment_month
- Final dataset: [71959] rows, [20] columns
- Overall no-show rate: ~20.2%
